```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load an image
# Code reference: https://docs.opencv.org/2.4/doc/tutorials/introduction/display_image/display_image.html
image = cv2.imread('images/whu_klein.jpg', cv2.IMREAD_COLOR)
```

```python
plt.imshow(image)
```

```python
# BGR is open cv's default color order, change it to rgb which is used by matplotlib
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Create a kernel to create a blurry image
kernel_b = np.array([[1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]])
print(kernel_b)
```

```python
# Apply the kernel
# Code reference: https://docs.opencv.org/2.4/modules/imgproc/doc/filtering.html#void%20filter2D(InputArray%20src,%20OutputArray%20dst,%20int%20ddepth,%20InputArray%20kernel,%20Point%20anchor,%20double%20delta,%20int%20borderType)
blurry = cv2.filter2D(src=rgb_image, kernel=kernel_b, ddepth=-1)
plt.imshow(blurry)
```

```python
# Create a kernel to extract vertical lines
kernel_v = np.array([[-1, 2, -1],
                   [-1, 2, -1],
                   [-1, 2, -1]])
print(kernel_v)
```

```python
# Apply the kernel
vertical = cv2.filter2D(src=rgb_image, kernel=kernel_v, ddepth=-1)
plt.imshow(vertical)
```

```python
# Create a kernel to extract horizontal lines
kernel_h = np.array([[-1, -1, -1],
                   [2, 2, 2],
                   [-1, -1, -1]])
print(kernel_h)
```

```python
# Apply the kernel
horizontal = cv2.filter2D(src=image, kernel=kernel_h, ddepth=-1)
plt.imshow(horizontal)
```
